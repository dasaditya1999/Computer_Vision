```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
image = cv2.imread('Ashok_Chakra.jpg.png')

plt.imshow(image)
```

```python
img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
pic_HSL = cv2.cvtColor(img, cv2.COLOR_RGB2HLS)
plt.imshow(pic_HSL)
```

```python
pic_HSV = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
plt.imshow(pic_HSV)
```

```python
pic_HSL_ = cv2.cvtColor(image, cv2.COLOR_RGB2HLS)
plt.imshow(pic_HSL_)
```

```python
pic_HSV_ = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)
plt.imshow(pic_HSV_)
```
