```python
# matplotlib --> RGB
# OpenCV ---> BGR
# In order to get original image, we need to convert the BGR of OpenCV to RGB. Matplotlib expects the image order to be RGB.
```

```python
correct_pic = cv2.cvtColor(picture, cv2.COLOR_BGR2RGB)
```

```python
# correct_pic
```

```python
plt.imshow(correct_pic)
```

```python
correct_pic.shape
```

```python
resize_image = cv2.resize(correct_pic,(1000,500))
plt.imshow(resize_image)
```

```python
new_image = cv2.resize(correct_pic,(0,0),correct_pic,0.5,0.8)
new_image.shape
```

```python
plt.imshow(new_image)
```

```python
flipped_1 = cv2.flip(correct_pic,-1)
plt.imshow(flipped_1)
```

```python
cv2.imwrite('totally_new.jpg', correct_pic)
```
