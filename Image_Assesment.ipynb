```python
# Computer-Vision-with-Python/Computer-Vision-with-Python/DATA/dog_backpack.jpg

import cv2
import matplotlib.pyplot as plt
%matplotlib inline

image = cv2.imread('Computer-Vision-with-Python/Computer-Vision-with-Python/DATA/dog_backpack.jpg')

corrected_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image)
```

```python
new_image = corrected_image.copy()
```

```python
plt.imshow(corrected_image)
```

```python
flipped_img = cv2.flip(corrected_image, 0)
plt.imshow(flipped_img)
```

```python
cv2.rectangle(corrected_image,(245,380),(610,760), color=(255,0,0), thickness=-1)

plt.imshow(corrected_image)
```

```python
import numpy as np
vertices = np.array([[480,350],[300,710],[610,710]], dtype=np.int32)
points = vertices.reshape((-1,1,2))
cv2.polylines(new_image, [points], isClosed=True, color=(0,0,255), thickness = 10)
plt.imshow(new_image)
```

```python
import numpy as np
vertices = np.array([[480,350],[300,710],[610,710]], dtype=np.int32)
points = vertices.reshape((-1,1,2))
cv2.fillPoly(new_image, [points], color=(0,0,255))
plt.imshow(new_image)
```
