<b>Drawing a box by clicking on the video</b>

```python
import cv2
import time

pt1 = (0,0)
pt2 = (0,0)
drawing = False


def draw_box(event, x,y,flags,params):
    
    global pt1,drawing,pt2
    
    if drawing:
        pt1 = (0,0)
        pt2 = (0,0)
#     print(event)
    if event==cv2.EVENT_LBUTTONDOWN:
#         if drawing==False:
            drawing=True
            pt1=(x,y)
            xn = x+270
            yn = y+400
            pt2 = (xn,yn)

capture = cv2.VideoCapture(0)        

height = capture.get(cv2.CAP_PROP_FRAME_HEIGHT)
width = capture.get(cv2.CAP_PROP_FRAME_WIDTH)

cv2.namedWindow('drawing on video')
        
cv2.setMouseCallback('drawing on video', draw_box)

while True:
    ret,frame = capture.read()
    
#     frame = cv2.resize(frame, (0,0), frame, 0.4, 0.3)

    if drawing:
        cv2.rectangle(frame, pt1, pt2, (0, 0, 255), 2)
        
    cv2.imshow('drawing on video', frame)
    
    if cv2.waitKey(20) & 0XFF==27:
        break

capture.release()        
cv2.destroyAllWindows()
```

<b>Drawing a box by doing the hardcoding</b>

```python
import cv2
import time

capture = cv2.VideoCapture(0)

height = int(capture.get(cv2.CAP_PROP_FRAME_HEIGHT))
width = int(capture.get(cv2.CAP_PROP_FRAME_WIDTH))

while capture.isOpened():
    
    ret,frame = capture.read()
    
#     frame = cv2.resize(frame, (0,0), frame, 0.4, 0.3)
    
    cv2.rectangle(frame, (width//2,height//2), ((width//2)+100,(height//2)+100), color=(255,0,0), thickness=6)
    
    cv2.imshow('drawing on video', frame)
    
    if cv2.waitKey(20) & 0XFF==27:
        break
        
cv2.destroyAllWindows()
capture.release()
```
