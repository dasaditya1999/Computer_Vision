```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

image = cv2.imread('Computer-Vision-with-Python/Computer-Vision-with-Python/DATA/sammy.jpg')
puppy = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(puppy)
```

```python
puppy_face = cv2.imread('Computer-Vision-with-Python/Computer-Vision-with-Python/DATA/sammy_face.jpg')
puppy_face = cv2.cvtColor(puppy_face, cv2.COLOR_BGR2RGB)
plt.imshow(puppy_face)
```

```python
method = eval('cv2.TM_CCOEFF')
result = cv2.matchTemplate(puppy,puppy_face,method)
plt.imshow(result)
```

```python
copy_puppy = puppy.copy()
plt.imshow(copy_puppy)
```

```python
methods = ['cv2.TM_CCOEFF', 'cv2.TM_CCOEFF_NORMED', 'cv2.TM_CCORR','cv2.TM_CCORR_NORMED', 'cv2.TM_SQDIFF', 'cv2.TM_SQDIFF_NORMED']

height,width, dim = puppy_face.shape
# top_left = (0,0)
# right_bottom = (0,0)

for m in methods:
    copy_puppy = puppy.copy()
    method = eval(m)
    result = cv2.matchTemplate(copy_puppy, puppy_face,method)
    
    min_value,max_value,min_loc,max_loc = cv2.minMaxLoc(result)
    
    if method in [cv2.TM_SQDIFF,cv2.TM_SQDIFF_NORMED]:
        top_left = min_loc
    else:
        top_left = max_loc

# Min_value is the value, matching has happedned very less
# Max_value is the value, matching has happedned very high
# Max_loc is the location, where the template matches the most
# Min_loc is the location, where the template matches the least

    print(max_loc)
    print(max_loc[0]+width,max_loc[1]+height)
    
    right_bottom = (top_left[0]+width, top_left[1]+height)
    
    cv2.rectangle(copy_puppy,top_left,right_bottom,(0,255,0),7)
    
    plt.suptitle(m)
    plt.subplot(121)
    plt.imshow(result)
#     plt.show()
    
    plt.subplot(122)
    plt.imshow(copy_puppy)
    plt.show()
```
