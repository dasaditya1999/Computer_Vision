```python
image = cv2.imread('Computer-Vision-with-Python/Computer-Vision-with-Python/DATA/crossword.jpg', 0)
plt.imshow(image, cmap='gray')
```

```python
def show_pic(image):
    fig = plt.figure(figsize=(15,15))
    ax = fig.add_subplot(111)
    ax.imshow(image, cmap='gray')
```

```python
output = show_pic(image)
```

```python
# plt.imshow(output)
```

```python
ret1, image_binary = cv2.threshold(image, 190, 255, cv2.THRESH_BINARY)
output = show_pic(image_binary)
```

```python
# ret2, adaptive_img = cv2.threshold(image, )
adaptive_image = cv2.adaptiveThreshold(image,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,19,10)
show_pic(adaptive_image)
```

```python
blended = cv2.addWeighted(src1=image_binary, alpha =0.5, src2=adaptive_image, beta=0.5, gamma=0)
show_pic(blended)
```
