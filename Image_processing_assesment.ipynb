```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display_image(image,cmap=None):
    fig = plt.figure(figsize=(10,10))
    ax = fig.add_subplot(111)
    ax.imshow(image, cmap)
```

```python
image = cv2.imread('Computer-Vision-with-Python/Computer-Vision-with-Python/DATA/giraffes.jpg')
image_RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
display_image(image_RGB)
```

```python
image = cv2.imread('Computer-Vision-with-Python/Computer-Vision-with-Python/DATA/giraffes.jpg',0)
plt.imshow(image, cmap='gray')
```

```python
th1, binary_thresh = cv2.threshold(image,100, 255,cv2.THRESH_BINARY)
```

```python
display_image(binary_thresh,cmap='gray')
```

```python
image = cv2.imread('Computer-Vision-with-Python/Computer-Vision-with-Python/DATA/giraffes.jpg')
image_HSV = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
display_image(image_HSV)
```

```python
image = cv2.imread('Computer-Vision-with-Python/Computer-Vision-with-Python/DATA/giraffes.jpg',0)
plt.imshow(image, cmap='gray')
```

```python
sobel_x = cv2.Sobel(image, cv2.CV_64F, dx=1,dy=0,ksize =5)
display_image(sobel_x,cmap='gray')
```

```python
sobel_y = cv2.Sobel(image, cv2.CV_64F, dx=0,dy=1,ksize=5)
display_image(sobel_y,cmap='gray')
```

```python
sobel_xy = cv2.Sobel(image, cv2.CV_64F, dx=1,dy=1,ksize=5)
display_image(sobel_xy,cmap='gray')
```

```python
image = cv2.imread('Computer-Vision-with-Python/Computer-Vision-with-Python/DATA/giraffes.jpg')
show_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
display_image(show_image)
```

```python
hist_value = cv2.calcHist([image],channels=[0],mask=None, histSize=[256],ranges=[0,256])
hist_value
```

```python
plt.plot(hist_value)
```

```python
color=('b','g','r')
for i,col in enumerate(color):
    hist_value = cv2.calcHist([image],channels=[i],mask=None,histSize=[256], ranges=[0,256])
    plt.plot(hist_value,col)
plt.title('COLOR HISTOGRAM')
plt.xlabel('0 to 255 Color values')
plt.ylabel('No of pixels')
plt.show()
```
