```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dark_horse = cv2.imread('Computer-Vision-with-Python/Computer-Vision-with-Python/DATA/horse.jpg')
show_horse = cv2.cvtColor(dark_horse, cv2.COLOR_BGR2RGB)

blue_bricks = cv2.imread('Computer-Vision-with-Python/Computer-Vision-with-Python/DATA/bricks.jpg')
show_bricks = cv2.cvtColor(blue_bricks, cv2.COLOR_BGR2RGB)

rainbow = cv2.imread('Computer-Vision-with-Python/Computer-Vision-with-Python/DATA/rainbow.jpg')
show_rainbow =  cv2.cvtColor(rainbow, cv2.COLOR_BGR2RGB)
```

```python
hist_value = cv2.calcHist([dark_horse], channels=[0], mask=None, histSize = [256], ranges=[0,256])
```

```python
plt.plot(hist_value, color='g')
```

```python
hist_value = cv2.calcHist([blue_bricks], channels=[2], mask=None, histSize = [256], ranges=[0,256])
hist_value.shape
```

<b>PLOTTING THREE COLOR HISTOGRAM<b>

```python
img = blue_bricks
color = ('b','g','r')
for i,col in enumerate(color):
    print(i, col)
    histr = cv2.calcHist([img], channels=[i], mask=None, histSize=[256], ranges=[0,256])
    plt.plot(histr, color=col)
    plt.xlim([0,256])
#     plt.ylim([0,7500])
plt.xlabel('0 to 255')
plt.ylabel('No of Pixels having respective value from 0 to 255')
plt.title('Image')
plt.show()
```

```python
img = blue_bricks
color = ('b','g','r')
for i,col in enumerate(color):
    hist = cv2.calcHist([img], channels = [i], mask=None, histSize=[256], ranges=[0,256])
    plt.plot(hist, color=col)
plt.show()
```

```python
img = dark_horse
color = ('b','g','r')
for i,col in enumerate(color):
    print(i, col)
    histr = cv2.calcHist([img], channels=[i], mask=None, histSize=[256], ranges=[0,256])
    plt.plot(histr, color=col)
    plt.xlim([0,56])
    plt.ylim([0,500000])
plt.title('Image')
plt.xlabel('0 to 255')
plt.ylabel('No of Pixels having respective value from 0 to 255')
plt.show()
```

```python
img = rainbow
color = ('b','g','r')
for i,col in enumerate(color):
    print(i, col)
    histr = cv2.calcHist([img], channels=[i], mask=None, histSize=[256], ranges=[0,256])
    plt.plot(histr, color=col)
    plt.xlim([0,106])
plt.title('Image')
plt.xlabel('0 to 255')
plt.ylabel('No of Pixels having respective value from 0 to 255')
plt.show()
```
